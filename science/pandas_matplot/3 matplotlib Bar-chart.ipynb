### 1. Core Purpose: Why Use a Bar Chart? 

Bar charts visually compare discrete categories by encoding data values as the length (or height) of rectangular bars. Their primary goals are: 

   - Comparison: Quickly show differences in magnitude between distinct groups (e.g., sales by product, votes by candidate).
   - Ranking: Easily identify the largest/smallest values (e.g., top 10 countries by GDP).
   - Distribution: Reveal patterns in categorical data (e.g., survey responses: "Strongly Agree" to "Strongly Disagree").
   - Part-to-Whole Relationships (with stacked variants): Show how sub-groups contribute to a total (e.g., market share by region within a country).

```python
product = ['PS4', 'PS3','PS5', 'black coffe','Lattie','sperso','Tea']

soled_1 = [200,150,350,200,240,360,401]
soled_2 = [210,240,400,240,300,450,300]
```

```python
plt.figure(figsize=(12,5))

plt.bar(product , soled_1, label='store 1')
plt.bar(product , soled_2, bottom=soled_1,  label='store 2')


plt.title('store-1 sales')
plt.grid()

plt.legend()
```
