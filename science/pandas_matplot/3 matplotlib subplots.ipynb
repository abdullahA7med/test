```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

the sec way and petter way 

- name of figure , axis name = plt.subplots(numbers of rows , numbers of columns)

```python
productes = ['apple','juci','PS5','PS4','black coffe','meat','Tv', 'nuts','shapo','wepons']

counts = [100,120,90,150,70,200,40,90,120,150]
```

```python
plt.figure(figsize = (20,5))

fig , ax = plt.subplots(2,1)

ax[0].bar(productes, counts,color='r')
ax[0].set_title('frequency product')
ax[0].tick_params(axis='x', labelrotation=45)


ax[1].hist(counts)

# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
