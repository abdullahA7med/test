### üîç Core Purpose of a Line Plot 

**A line plot visualizes continuous change or trends across a sequential, ordered dimension (most commonly time). Its primary goals are:**

   - Reveal Trends: Show upward/downward movements, cycles, or patterns (e.g., growth, seasonality, decay).
   - Highlight Continuity: Emphasize that data points are connected (e.g., measurements taken at regular intervals).
   - Compare Series: Overlay multiple lines to compare trends across categories (e.g., sales of Product A vs. B).
   - Forecast/Extrapolate: Extend lines to predict future values (often with confidence intervals).
   - Show Rate of Change: Steep slopes = rapid change; flat lines = stability.

**can add more than one plot at same figure.**

```python

year_25 = ['January','February','March','April','May','June','July','August','September','October','November','December']

coffe_shop = [120,40,90,110,200,150,901,188,189,200,110,200]

ps_center = [100,90,189,200,72,90,40,80,240,120,150,200]
```

```python
plt.figure(figsize=(15,6))
plt.plot(year_25 , coffe_shop, label='coffe shop', marker='^')
plt.plot(year_25 , ps_center, label='PS center', marker='o')

plt.title('earns coffe shop /VS/ Ps center')
plt.xlabel('months')
plt.ylabel('earns')


plt.legend() 
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(year_25 , coffe_shop, label='coffe shop', marker='^' , ls='--')
plt.plot(year_25 , ps_center, label='PS center', marker='o', linestyle=':')

plt.title('earns coffe shop /VS/ Ps center')
plt.xlabel('months')
plt.ylabel('earns')


plt.legend() 
plt.grid()

plt.show()
```
