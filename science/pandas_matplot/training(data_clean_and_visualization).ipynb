```python
dic = {
'Python': 21860,
'R': 5334,
'SQL': 10756,
'C' :4709,
'C++': 5535,
'Java': 4769,
'Javascript': 4332,
'Julia': 305,
'Swift' :242,
'Bash' :2216,
'MATLAB': 2935}

Q7_series=pd.Series(dic)
Q7_series = Q7_series.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,9))
plt.pie(Q7_series, labels=Q7_series.index, autopct="%1.1f%%",)

plt.xticks(rotation=45)
# plt.legend()
plt.show()
```

```python
cust_color = ["#03045e","#023e8a","#0077b6","#0096c7","#00b4d8","#48cae4","#90e0ef","#ade8f4","#caf0f8"]
```

```python
fig , ax = plt.subplots(1,2, figsize=(20,8))
fig.patch.set_facecolor('#F2F2F2')


ax[0].bar(Q7_series.index , Q7_series.values, color=cust_color)
ax[0].set_ylabel('Count')
ax[0].set_xlabel('programing Lang')
ax[0].tick_params(labelrotation=45, labelsize=14)
# ax[0].patch.set_facecolor('#F2F2F3')
# sec plot

ax[1].pie(Q7_series, labels=Q7_series.index, autopct="%1.1f%%", colors=cust_color , explode=[.04 for i in range(len(Q7_series))])


plt.suptitle('programing language',size=30, weight='bold', fontname = 'monospace')
plt.show()
```
