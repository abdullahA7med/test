### 3 Core Purpose: What a Histogram Really Does 

    Visualize Distribution: Shows how numerical data is spread across intervals (bins), revealing patterns invisible in raw data.
   
   - Identify Key Characteristics:
      
       - Shape: Is data symmetric (bell-shaped), skewed left/right, uniform, or bimodal?
       - Central Tendency: Where is the "peak" (mode)? How does it relate to mean/median?
       - Spread/Variability: How wide is the distribution? Are values clustered or dispersed?
       - Outliers: Extreme values sticking out from the main body.
       - Gaps: Unusual empty intervals suggesting data anomalies.
         
    Test Assumptions: Check if data follows a normal distribution (critical for many statistical tests).
    Compare Distributions: Overlay histograms (or use facets) to compare groups (e.g., test scores from different classes).
     

```python
ages = [10,10,15,20,20,25,25,26,18,10,15,20,20,25,30,22,10]

plt.figure(figsize=(12,5))
plt.hist(ages, bins=7, edgecolor='black')


plt.show()
```

```python
ages = [10,10,15,20,20,25,25,26,18,10,15,20,20,25,30,22,10]
bins = [10,15,20,25,30]
plt.figure(figsize=(12,5))
plt.hist(ages, bins=bins, edgecolor='black')


plt.show()
```
