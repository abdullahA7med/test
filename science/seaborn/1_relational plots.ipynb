```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
flight = sns.load_dataset('flights')
flight
```

```python
flight['year'].value_counts()
```

```python
flight.month.value_counts()
```

```python
feb_flight = flight.loc[flight['month'] == 'Feb']
feb_flight
```

```python
sns.lineplot(x='year', y='passengers', data=feb_flight)
```

```python
# as we know we can use matplotlib with it

sns.lineplot(x='year', y='passengers', data=feb_flight)

plt.suptitle('passengers per year')
plt.show()
```

```python
sns.lineplot(x='year', y='passengers', data=flight)
```

```python
# but need every month have own plot.
# use hue, to devide theme.

sns.lineplot(x='year', y='passengers', data=flight,hue='month')
plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.lineplot(x='year', y='passengers', data=flight,hue='month', style='month')
plt.show()
```

#### test anouther data.

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event')
```

```python
# to make more than one plot use relplot , can devide them as you wish.

sns.relplot(kind='line',x='timepoint', y='signal', data=fmri, hue='event', style='region')
```

```python
# using col param , to devide theme as any column.

sns.relplot(kind='line',x='timepoint', y='signal', data=fmri, hue='event', style='region', col='event')
```

```python
sns.relplot(kind='line',x='timepoint', y='signal', data=fmri, hue='event', style='region', col='event', row='region')
```
