```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

**let's know scatter plot in relational plots**

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,hue='sex')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,hue='sex',style='time')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,hue='sex',style='time',size='smoker')
plt.show()
```

```python
# change the color of points.
# use palette.
plt.figure(figsize=(10,8))
sns.scatterplot(x='total_bill', y='tip', data=tips,hue='sex',style='time' ,palette=('coolwarm'))
plt.show()
```

**to customize your markes ,use marker param.**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='sex' ,markers={'Male':'D', 'Female':'d'})
```

**using realplot**

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='time', style='sex', markers={'Male':'D' , 'Female':'d'},
           col='sex', row='time')

plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='time', style='sex', markers={'Male':'D' , 'Female':'d'},
           col='sex', row='day')

plt.show()
```
