```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

**have one plot.**

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()
```

can plot relation between categorical and numerical.

```python
# used x_jitter to add space 

sns.regplot(x='size', y='tip', data=tips, x_jitter=0.1)
plt.show()
```

**some time i need plot the relation but using mean**

```python
sns.regplot(x='size', y='tip', data=tips, x_estimator=np.mean )
plt.show()
```

## adva -> implot

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='time',row='smoker')
plt.show()
```
