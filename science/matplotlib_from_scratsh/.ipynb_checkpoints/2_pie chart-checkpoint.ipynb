```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
skills = ['python','js','Network','linux Os','pentest knowledge']

percentage = [90,50,60,85,55]
```

```python
plt.pie(percentage, labels=skills)
```

```python
custom_color = ["#ffcbf2","#f3c4fb","#ecbcfd","#e5b3fe","#e2afff","#deaaff","#d8bbff","#d0d1ff","#c8e7ff","#c0fdff"]
```

```python
plt.pie(percentage, labels=skills)
plt.show()
```

```python
# to add percentage.


plt.pie(percentage, labels=skills, autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(percentage, labels=skills, autopct='%0.0f%%')
plt.show()
```

```python
plt.pie(percentage, labels=skills, autopct='%0.0f%%',explode=[0.1,0.01,0.05,0.1,0.009])
plt.show()
```

```python
plt.pie(percentage, labels=skills, autopct='%0.0f%%',explode=[0.1,0.01,0.05,0.1,0.009], wedgeprops={'edgecolor':'black','lw':2})
plt.show()
```

```python
plt.pie(percentage, labels=skills, autopct='%0.0f%%',explode=[0.1,0.01,0.05,0.1,0.009],
        wedgeprops={'edgecolor':'black','lw':2},
       counterclock=False)
plt.show()
```

```python
# can change the color.


plt.pie(percentage, labels=skills, autopct='%0.0f%%',explode=[0.1,0.01,0.05,0.1,0.009],colors=custom_color,
        wedgeprops={'edgecolor':'black','lw':2},
       counterclock=False)
plt.show()
```
