```python
z = ['a','b','c','d','e','f','g']
c = [11,1,34,12,4,5,6]
```

```python
plt.plot(z,c)
```

### using attributes:

```python
# marker

plt.plot(z,c ,marker='.',markersize=12)
```

```python
# markersize / or / ms 
# used to change size of marker

plt.plot(z,c ,marker='.',markersize=12)
```

```python
# markerfacecolor / mfc 
# jop > change color of marker

plt.plot(z,c ,marker='.',markersize=12, mfc='black')
```

```python
# linestyle /or/ ls 
# change style of the line

plt.plot(z,c ,marker='.',markersize=12, mfc='black' , ls='-.')
```

```python
# color
# cutomize the color

plt.plot(z,c ,marker='.',markersize=12, mfc='black' , ls='-.', color='c')
```

```python
# multible lines at same one

days = np.arange(1,8)

luxor_Tem = [45,55,43,42,47,48,50]
cairo_Tem = [40,45,42,46,41,39,41]
qena_Tem  = [45,54,55,43,46,47,51]
```

```python
plt.plot(days , luxor_Tem, marker='.', mfc='red', ms=12, label='luxor', ls='--')
plt.plot(days , cairo_Tem, marker='^', ms=12 , mfc='black', label='cairo', ls='-.')
plt.plot(days , qena_Tem, marker='s', ms=12 , mfc='c', label='Qena', ls=':')


plt.legend(bbox_to_anchor=(1.2,1))
```

```python
 plt.plot(days , luxor_Tem, marker='.', mfc='red', ms=12, label='luxor', ls='--')
plt.plot(days , cairo_Tem, marker='^', ms=12 , mfc='black', label='cairo', ls='-.')
plt.plot(days , qena_Tem, marker='s', ms=12 , mfc='c', label='Qena', ls=':')


plt.xlabel('Days', fontdict = {'family':'serif',
                              'color': '#2F5249',
                              'size' : 16})

plt.ylabel('Tempretures', fontdict = {'family':'serif',
                              'color': '#2F5249',
                              'size' : 16})

plt.title('Tempretures per Days', fontdict= {'family' : 'monospace'} , color='#5682B1', fontsize=16)

plt.legend(bbox_to_anchor=(1.2,1))
plt.show()
```
