```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
subject = ['arabic','math','statistic','history','english','chemistry','physics']
scores = [1056,1903,1987,1206,1403,1302,1002]
```

```python
# the recommended graph is -> bar.

fig , ax = plt.subplots()

ax.bar(subject , scores)
```

```python
cus_color = ["#e63946","#f1faee","#a8dadc","#457b9d","#1d3557"]
```

```python
fig , ax = plt.subplots(1,2, figsize=(20,8))
fig.patch.set_facecolor('#F2F2F2')


ax[0].bar(subject , scores,color=cus_color,)
ax[0].set_title('distripution of score for subjects',fontsize=20)
ax[0].tick_params(labelrotation=45, labelsize=14)

# sec plot

ax[1].pie(scores,
          labels=subject,
          autopct='%0.0f%%',
          explode=[0.1,0.01,0.05,0.1,0.009,0.1,0.1],
          colors = cus_color,
         wedgeprops={'edgecolor':'black','lw':2},
         textprops={'fontsize': 18})

plt.suptitle('distipution',size=30, weight='bold', fontname = 'monospace')
plt.tight_layout()
plt.savefig('Test_vis.png')
plt.show()
```
